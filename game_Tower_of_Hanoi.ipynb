```python
class TowerOfHanoi:
    def __init__(self, num_disks, max_moves):
        self.num_disks = num_disks
        self.max_moves = max_moves
        self.towers = {
            'A': list(reversed(range(1, num_disks + 1))),  # Tiang awal dengan cakram
            'B': [],  # Tiang kosong
            'C': []   # Tiang tujuan
        }
        self.moves = 0

    def move_disk(self, from_tower, to_tower):
        if self.moves >= self.max_moves:
            print("Maksimum langkah tercapai!")
            return

        if not self.towers[from_tower]:
            print(f"Tiang {from_tower} kosong!")
            return

        if self.towers[to_tower] and self.towers[to_tower][-1] < self.towers[from_tower][-1]:
            print("Tidak bisa meletakkan cakram besar di atas yang lebih kecil!")
            return

        disk = self.towers[from_tower].pop()  # Pop dari tiang asal
        self.towers[to_tower].append(disk)  # Push ke tiang tujuan
        self.moves += 1
        print(f"Memindahkan cakram {disk} dari {from_tower} ke {to_tower}")
        self.display_towers()

    def display_towers(self):
        print("\nStatus Tiang:")
        for tower, disks in self.towers.items():
            print(f"{tower}: {disks}")
        print(f"Langkah: {self.moves}/{self.max_moves}\n")

# Input dua digit terakhir NPM sebagai jumlah langkah maksimal
npm_last2 = int(input("Masukkan dua digit terakhir NPM: "))

# Inisialisasi permainan dengan 3 cakram dan batas langkah
hanoi = TowerOfHanoi(num_disks=3, max_moves=npm_last2)
hanoi.display_towers()

# Gameplay manual
while hanoi.moves < hanoi.max_moves:
    from_tower = input("Pindahkan dari tiang (A/B/C): ").upper()
    to_tower = input("Pindahkan ke tiang (A/B/C): ").upper()

    if from_tower in "ABC" and to_tower in "ABC" and from_tower != to_tower:
        hanoi.move_disk(from_tower, to_tower)
    else:
        print("Input tidak valid, coba lagi!")

    if hanoi.towers['C'] == [3, 2, 1]:
        print("Selamat! Anda menyelesaikan permainan.")
        break
```
